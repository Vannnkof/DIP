```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread("cherno-belye-risunki-po-kletochkam-13.png", cv2.IMREAD_GRAYSCALE)
```

```python
img_color = cv2.imread("1.jpg", cv2.IMREAD_COLOR)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap='gray')
plt.show();
```

```python

print("BGR:\n")
plt.figure(figsize=(8,8))
plt.imshow(img_color)
plt.show();
print("RGB:\n")
plt.figure(figsize=(8,8))
img_rgb=cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show();
print("HSV:\n")
plt.figure(figsize=(8,8))
img_hsv=cv2.cvtColor(img_color,cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.show();
```

```python
print("Окремі канали RGB:")
plt.figure(figsize=(10, 10))
plt.subplot(131);
R = np.zeros(img_rgb.shape, dtype='uint8')
R[:, :, 0] = img_rgb[:, :, 0]
plt.imshow(R)
plt.subplot(132);
G = np.zeros(img_rgb.shape, dtype='uint8')
G[:, :, 1] = img_rgb[:, :, 1]
plt.imshow(G)
plt.subplot(133);
B = np.zeros(img_rgb.shape, dtype='uint8')
B[:, :, 2] = img_rgb[:, :, 2]
plt.imshow(B)
plt.show()

print("Окремі канали HSV:")
plt.figure(figsize=(10, 10))
plt.subplot(131);
R = np.zeros(img_hsv.shape, dtype='uint8')
R[:, :, 0] = img_hsv[:, :, 0]
plt.imshow(R)
plt.subplot(132);
G = np.zeros(img_hsv.shape, dtype='uint8')
G[:, :, 1] = img_hsv[:, :, 1]
plt.imshow(G)
plt.subplot(133);
B = np.zeros(img_hsv.shape, dtype='uint8')
B[:, :, 2] = img_hsv[:, :, 2]
plt.imshow(B)
plt.show()
```

```python
print("Область інтересу для ЧБ:\n")
img_crop = img_gray[0:int(img_gray.shape[0]/2), 0:int(img_gray.shape[1]/2)]
plt.imshow(img_crop,cmap = "gray")
plt.show()
```

```python
cv2.imwrite("laba_write.png", img_crop)
img_write = cv2.imread("laba_write.png", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(8,8))
plt.imshow(img_write,cmap='gray')
plt.show();
```
